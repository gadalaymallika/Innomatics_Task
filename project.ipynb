Is there a relationship between the College Tier (Tier 1, 2, 3) and the Salary earned by students?"

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# Boxplot to visualize the relationship between CollegeTier and Salary
plt.figure(figsize=(8,6))
sns.boxplot(x='CollegeTier', y='Salary', data=df)
plt.title('Salary Distribution by College Tier')
plt.xlabel('College Tier')
plt.ylabel('Salary')
plt.show()

# ANOVA test to check if CollegeTier significantly affects Salary
anova_result = stats.f_oneway(df[df['CollegeTier'] == 1]['Salary'],
                              df[df['CollegeTier'] == 2]['Salary'],
                              df[df['CollegeTier'] == 3]['Salary'])

print(f"ANOVA F-statistic: {anova_result.statistic}")
print(f"P-Value: {anova_result.pvalue}")
```
